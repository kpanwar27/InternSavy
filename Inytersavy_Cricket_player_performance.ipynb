```python
# Data Source = "https://www.kaggle.com/datasets/saivamshi/cricket-world-cup-2019-players-data"

batsman=pd.read_csv('/content/drive/MyDrive/Internship/Batsman_Data.csv')
bowler=pd.read_csv('/content/drive/MyDrive/Internship/Bowler_data.csv')
ground=pd.read_csv('/content/drive/MyDrive/Internship/Ground_Averages.csv')
odi_result=pd.read_csv('/content/drive/MyDrive/Internship/ODI_Match_Results.csv')
odi_totals=pd.read_csv('/content/drive/MyDrive/Internship/ODI_Match_Totals.csv')
wc=pd.read_csv('/content/drive/MyDrive/Internship/WC_players.csv')
```

```python
data = [batsman, bowler, ground, odi_result, odi_totals, wc]

# Define the file names corresponding to each DataFrame

file_paths = [
    '/content/drive/MyDrive/Internship/Batsman_Data.csv',
    '/content/drive/MyDrive/Internship/Bowler_data.csv',
    '/content/drive/MyDrive/Internship/Ground_Averages.csv',
    '/content/drive/MyDrive/Internship/ODI_Match_Results.csv',
    '/content/drive/MyDrive/Internship/ODI_Match_Totals.csv',
    '/content/drive/MyDrive/Internship/WC_players.csv'
]

for k, file_path in zip(data, file_paths):
    file_name = os.path.basename(file_path)
    print(f"File Name: {file_name}")

    # Print head
    print("\nFile_Content:")
    print(k.head(2))

    # Print info
    print("\nInfo:")
    print(k.info())

    # Print Shape
    print("\nShape:")
    print(k.shape)

    # Print Null
    print("\nNull_Values:")
    print(k.isnull().sum())

    # Print Duplicates
    print("\nTotal_Duplicates:")
    print(k.duplicated().sum())

    print("=" * 70)  # Separating line for clarity
```
